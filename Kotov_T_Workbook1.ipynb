```python
import numpy
import matplotlib.pyplot as plt

list_ = numpy.random.rand(50)
avg = numpy.mean(list_)
median = numpy.median(list_)

print(f"Среднее значение: {avg}\nМедианное значение: {median}")

if median > avg:
    print("Медианное значение больше среднего значения")
elif avg < median:
    print("Среднее значение больше медианного значения")
else:
    print("Среднее и медианное значения равны.")

plt.scatter(range(len(list_)), list_, color="green")
plt.grid(True)
plt.show()
```

```python
import math

def f(x):
    return (math.sqrt(1 + math.pow(math.e, math.sqrt(x)) + math.cos(math.pow(x, 2)))) / abs(1 - math.pow(math.sin(x), 3)) + math.log(2*x, math.e)

values = [f(i) for i in range(1, 11)]

plt.scatter(range(len(values) // 2), values[:len(values) // 2], color="green")
plt.grid(True)
plt.show()
plt.plot(range(len(values)), values, color="green")
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.integrate import nquad

import matplotlib.pyplot as plt

def f(x):
    return np.abs(np.cos(x*np.pow(np.e, np.cos(x)+np.log(x+1))))


x = np.arange(0, 10, 1)
y = f(x)

plt.plot(x, y, color="#ff7f0e")
plt.fill_between(x, y, color="#1f77b4")

area = nquad(f, [[0, 10]])
print("Площадь под графиком:", area[0])

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Chart")
plt.grid(True)
plt.show()
```

```python
from math import cos, sin, e, pow


def calculator(x, y, action):
    match (action):
        case "1":
            return x + y
        case "2":
            return x - y
        case "3":
            return x * y
        case "4":
            return x / y
        case "5":
            return pow(e, x + y)
        case "6":
            return sin(x + y)
        case "7":
            return cos(x + y)
        case "8":
            return pow(x, y)
    return 0


x = float(input("введите число 1:"))
action = input("""
введите число от 1 до 8 для выбора операции:
    1. Сложение
    2. Вычитание
    3. Умножение
    4. Деление
    5. Возведение e в степень суммы
    6. Синус
    7. Косинус
    8. Возведение в степень
""")
y = float(input("введите число 2:"))


print(calculator(x, y, action))
```
